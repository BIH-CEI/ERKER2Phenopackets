```python
from ERKER2Phenopackets.src.utils.io import read_files, phenopacket2dict

from pathlib import Path
```

```python
path = Path(r'C:\Users\Surface\OneDrive\Documents\DataSpell\ERKER2Phenopackets\ERKER2Phenopackets\data\out\phenopackets\example-phenopackets-from-synthetic-data')
```

```python
phenopackets = read_files(path)
```

```python
from ERKER2Phenopackets.src.analysis.tree_comparison.edit_dist import compare_structure
```

```python
import numpy as np

A = np.zeros((len(phenopackets), len(phenopackets)))
block = 'interpretations'#'phenotypicFeatures'
for i, p in enumerate(phenopackets):
    for j, q in enumerate(phenopackets):
        try:
            comparison = compare_structure(phenopacket2dict(p)[block], phenopacket2dict(q)[block])
        except KeyError:
            print(p.id, q.id, False)
            A[i, j] = 0
            continue
        else:
            print(p.id, q.id, comparison[0])
            A[i, j] = comparison[0]
            
```

```python
print(A)
```

```python
print(f'N: {len(phenopackets) ** 2}')
print(f'Matches: {np.count_nonzero(A) - len(phenopackets)}')
print(f'No matches: {np.count_nonzero(A == 0) + len(phenopackets)}')
```
